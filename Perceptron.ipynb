```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=100,n_features=3,n_informative=2,
                           n_redundant=0, n_repeated=0, n_clusters_per_class=1, 
                           class_sep=2,random_state=93)
# https://queirozf.com/entries/scikit-learn-examples-making-dummy-dataset
```

```python
def step(value):
  return 1 if value>=0 else 0

def basic_perceptron(X,y):
  X_new = np.insert(X,0,1, axis=1)
  w = np.ones(X_new.shape[1])
  lr = 0.01
  for i in range(10000):
    sample = np.random.randint(100)
    y_hat = step(np.dot(X_new[sample],w))
    w = w + lr*(y[sample]-y_hat)*X_new[sample]
  return w

```

```python
w = basic_perceptron(X,y)
```

```python
w
```

```python
tmp = np.linspace(3,-3)
x,y_plane = np.meshgrid(tmp,tmp)
plt.scatter(x,y_plane)

z = (-w[0]-w[1]*x -w[2]*y_plane) / w[3]
# https://towardsdatascience.com/support-vector-machines-svm-clearly-explained-a-python-tutorial-for-classification-problems-29c539f3ad8
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(X[:,0], X[:,1], X[:,2],c=y)
plt.title("simple 3D scatter plot")
ax.plot_surface(x, y_plane, z)
ax.view_init(25, 60)

# show plot
plt.show()
```
